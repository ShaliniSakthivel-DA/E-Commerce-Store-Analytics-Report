```python
# Case Study – Student Data Management System

student = {
            'name' : ['Shalini','Arun','Ram'],
            'age' : [18,19,20],
            'course' : ["Data Science", "AI", "Machine Learning"]
}
print(student)

# Add one new student

student['name'].append("Ravi")
student["age"].append(24)
student["course"].append("Python")

# Display all data

print("All Names:", student["name"])
print("All Ages:", student["age"])
print("All Courses:", student["course"])

# First student details

print("\n----- First Student -----")
print("Name:", student["name"][0])
print("Age:", student["age"][0])
print("Course:", student["course"][0])

print("\n----- Second Student -----")
print("Name:", student["name"][1])
print("Age:", student["age"][1])
print("Course:", student["course"][1])

print("\n----- New Student -----")
print("Name:", student["name"][3])
print("Age:", student["age"][3])
print("Course:", student["course"][3])
```

# Set

```python
# Case Study – Student Course Registration System using Set

# Creating set
course = {"Python", "SQL", "Power BI", "Python", "Excel"}
print(course)

# Add() course

course.add('Tableau')
print("\nAfter Adding New Course: ",course)

# update() - Add multiple courses

course.update(["Python","Java"])
print("\nAfter update multiple courses: ",course)

# remove() - Remove course

course.remove("Python")
print("\nAfter Removing Python Course: ",course)

# discard() - Remove safely

course.discard("C++")
print("\nAfter discard: ", course)

# Display total courses

print("\nTotal Courses: ",len(course))

# List of courses

print("\n---------Couses-----------")
print(course)
```
