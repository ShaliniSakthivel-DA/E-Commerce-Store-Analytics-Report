```python
#Set operations

A = {'Red', 'Green' , 'Yellow'}
B = {'Yellow', 'Orange', 'Red'}

print("A:", A)
print("B:", B)

# Union: All elements from both sets
print("Union:", A.union(B))

# Intersection: Common elements
print("Intersection:", A.intersection(B))

# Difference: Elements in A but not in B
print("Difference (A-B):", A.difference(B))
print("Difference (B-A):", B.difference(A))

# Symmetric Difference: Elements in A or B but not both
print("Symmetric Difference:", A.symmetric_difference(B))
```

```python
# Set Relations
X = {1, 2}
Y = {1, 2, 3, 5, 4}


# Checks if set is subset of another
print("X is subset of Y:", X.issubset(Y))
print("Y is subset of X:", Y.issubset(X))


# Checks if set is superset of another
print("X is superset of Y:", X.issuperset(Y))
print("Y is superset of X:", Y.issuperset(X))

Z = {2, 1}
# Checks if sets have no elements in common
print("X is disjoint with Z:", X.isdisjoint(Z))
```
