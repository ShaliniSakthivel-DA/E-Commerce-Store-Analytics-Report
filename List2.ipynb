```python
# Creating list

cart = ['milk', 'bread', 'eggs', 'apple']

cart1 = cart.copy()
print(cart)

#extend

cart.extend(['banana','butter'])
print(cart)

#insert

cart.insert(2,'cheese')
print(cart)

#remove

cart.remove('apple')
print(cart)

# pop()

cart.pop()
print(cart)
```

```python
# Index

cart = ['milk', 'bread', 'eggs', 'apple']
cart.index('eggs')

# count()

cart.count('milk')

# display first item and last item

cart = ['milk', 'bread', 'eggs', 'apple']
cart.index("milk")

cart = ['milk', 'bread', 'eggs', 'apple']
cart.index("apple")

# Display first 3 items from the list

cart [0:3]

# Find total number of items using len()

len(cart)

# Display all outputs using print()

print(cart)
```

```python
mark = list(map(int,input("Enter the marks: ").split(',')))
print(mark)

Total_Marks = sum(mark)
print("Total Mark:", Total_Marks)

Avg_Mark = Total_Marks / len(mark)
print("Avg Mark:", Avg_Mark)

Percentage = Avg_Mark / 100
print("Total Percentage: ",Percentage )

print(f"maximum mark: {max(mark)}")
print(f"Minimum mark: {min(mark)}")
```
